## Constant Function Oracle

### Constant Oracle (f(x)=0)

```python
from qiskit import QuantumCircuit

def constant_oracle():
    qc = QuantumCircuit(2)
    # Do nothing → always 0
    return qc
```

## Balanced Function Oracle

```python
def balanced_oracle():
    qc = QuantumCircuit(2)
    qc.cx(0,1)  # flips output if input = 1
    return qc
```

### Deutsch–Jozsa Circuit

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

def deutsch_jozsa(oracle):
    qc = QuantumCircuit(2,1)

    # Step 1: Initialize
    qc.x(1)  # output qubit = |1⟩

    # Step 2: Hadamard
    qc.h([0,1])

    # Step 3: Oracle
    qc.compose(oracle(), inplace=True)

    # Step 4: Hadamard on input
    qc.h(0)

    # Step 5: Measurement
    qc.measure(0,0)

    return qc
```

```python
qc_const = deutsch_jozsa(constant_oracle)

sim = AerSimulator()
result = sim.run(qc_const, shots=1024).result()
counts_const = result.get_counts()

print("Constant Function:", counts_const)
plot_histogram(counts_const)
plt.show()
```

```python
qc_bal = deutsch_jozsa(balanced_oracle)

result = sim.run(qc_bal, shots=1024).result()
counts_bal = result.get_counts()

print("Balanced Function:", counts_bal)
plot_histogram(counts_bal)
plt.show()
```
