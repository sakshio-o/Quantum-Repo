## Classical

```python
import numpy as np
from math import gcd

N = 15
a = 2  # chosen coprime

def classical_post_processing(r):
    if r % 2 != 0:
        return None
    
    x = pow(a, r//2, N)
    
    factor1 = gcd(x - 1, N)
    factor2 = gcd(x + 1, N)
    
    return factor1, factor2
```

## Quantum Period Finding Circuit

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

def shor_demo_circuit():
    qc = QuantumCircuit(4, 3)

    # Step 1: Superposition
    qc.h([0,1,2])

    # Step 2: Simulated modular exponentiation
    # (Simplified for teaching/demo)
    qc.cx(0,3)
    qc.cx(1,3)
    qc.cx(2,3)

    # Step 3: Inverse QFT (approximation)
    qc.h(0)
    qc.h(1)
    qc.h(2)

    # Measurement
    qc.measure([0,1,2], [0,1,2])

    return qc
```

```python
qc = shor_demo_circuit()
qc.draw('mpl')

sim = AerSimulator()
result = sim.run(qc, shots=1024).result()
counts = result.get_counts()

print("Measurement Results:", counts)
plot_histogram(counts)
plt.show()
```
